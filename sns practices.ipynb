```python
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-1)*flip)
```

```python
sinplot(-1)
```

```python
sns.set_style('whitegrid')
sinplot()
```

```python
sns.set_style('ticks')
sinplot()
sns.despine(left=True,bottom=True)
#sns.despin is used to remove the line in the graph at the ticks
```
